```python
import pandas as pd
```

```python
data = pd.read_csv("artwork_sample.csv")
```

```python
data.head()
```

```python
#select only specific columns
data.filter(items = ["id", "artist"]) #select only 2 columns
```

```python
#we can select columns with values contain in column name
data.filter(like = "artist") #select all the column with 'artist' in column's name
```

```python
#we can use regex to get the case-insensitive values in column's name
data.filter(regex = "(?i)year") #return every column with year in column's name, ignore case (year, Year)
```

```python
#select row
data.filter(axis = 0, regex = "^100.$") #axis = 0: row, axis = 1: column (by default)
```
