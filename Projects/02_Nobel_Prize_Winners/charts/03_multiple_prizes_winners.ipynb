```python
#1. Import library and data
import pandas as pd
data = pd.read_csv("dataset.csv")
data.head()
```

```python
#2. Clean and transform data
```

```python
#3. Get the list of scientists/writers/activists (names, and job) who win Nobel prize MORE THAN 1.

#a. Find all unique category in catgory
data["category"].unique()
#b. scientists/writers/activists categories: 'Chemistry', 'Literature', 'Peace'

#create a new dataframe with 2 columns: category and full_names
nobel_winners = data.loc[:, ["category", "full_name"]]

#c. count how many time each winner win Nobel prize
nobel_winners = data.groupby(["full_name", "category"]).size().reset_index(name = "nobel prizes")
#group by full_name and category, then use size() to count values in each group of full_name-category
#reset index of the new column create by size() to get a new dataframe, assign it back to nobel_winners

#d. filter rows to get winner with nobel prizes > 1 and is scientists/writers/activists and all the rows
names = nobel_winners.loc[nobel_winners["category"].isin(['Chemistry', 'Literature', 'Peace']) 
                          & (nobel_winners["nobel prizes"] > 1)] #filter rows

#merge names with data to get the categories of each winner
names
```
