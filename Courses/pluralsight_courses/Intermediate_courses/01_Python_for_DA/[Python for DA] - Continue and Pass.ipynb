```python
normal_cars = ["honda", "toyota", "kia morning", "mecerdes", "huyndai"]
```

```python
for car in normal_cars: #the for loops executed in order
    #1
    if car == "kia morning": 
        # when condition is True, it will skip iterate "kia morning" value and iterate the next value
        continue
    #2
    print(car) #every values in normal_cars will be iterated, except kia morning
```

```python
# another example
for i in range(10):
    
    if i % 2 == 0: #this block of code will skip all the value % 2 == 0 in list range(10)
        continue
    
    print(i) # when a value is skip, it will skip all code after
```

```python
# another continue example
for i in range(10, 21):
    
    if i in range(14, 19): # skip all the value in range (14, 19)
        continue
    
    print(i)
```

```python
# pass
for i in range(10, 20):
    pass #this mean there no code in here, go on
```

```python
for i in range(10, 20):
    if i in range(14, 19):
        pass #this mean the value in range(14, 19) will not be executed
    else:
        print(i)
```
