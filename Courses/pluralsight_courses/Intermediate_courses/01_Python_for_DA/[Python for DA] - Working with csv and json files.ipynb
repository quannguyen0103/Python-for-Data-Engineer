```python
import os
```

```python
# create a directory name csv_files
os.makedirs("csv_files")
```

```python
open("./csv_files/random_csv.csv")
```

```python
csv_list =[] #create a list to contain csv content
#with block code will make sure random_csv file will close after this

with open("./csv_files/random_csv.csv", "r") as c: #open csv file
    for ran_csv in c: #append every values in csv files into csv_list
        csv_list.append(ran_csv.strip("\n"))
```

```python
csv_list #now the list contain all csv file values
```

```python
import csv #csv module
```

```python
#read all the values in random_csv file using csv.reader function
with open("./csv_files/random_csv.csv", "r") as f:
    csv_reader = csv.reader(f) #read all the value in csv file
    #the reader function extract csv file into multidimension
    for i in csv_reader: #loop through and print all values in csv
        print(i)
```

```python
# The best library to work with csv files is pandas
import pandas as pd
```
