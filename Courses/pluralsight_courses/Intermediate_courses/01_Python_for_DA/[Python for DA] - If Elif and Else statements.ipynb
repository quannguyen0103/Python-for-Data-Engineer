```python
#set up variable math score
math_score = 75

#if elif else statements
#if else else statement will execute from the top to bottom
#if elif else statements will check all the conditions in order,
    # and stop and print the message at the first condition is True.
if math_score < 30: 
    print("I'm suck at math")
elif math_score < 50: # if the if condition is not True, it will execute this 
    print("I'm kinda bad at math")
elif math_score < 80:
    print("Hey, I'm good at math!")
else: # can use elif < 100 to make it clearer
    print("I'm a genius!!!!!")
```
