```python
#range(): generate number in a specific range
range(1, 5) #this range contain 1, but not contain value 5
```

```python
#if range only contain 1 number, it will start from 0
r = range(10) #this range not contain value 10
print(r)
```

```python
tuple(r) #turn range into tuple
```

```python
list(r) #turn list into tuple
```

```python
list(range(-4, 10))
```

```python
# range function in for loops
for i in range(10):
    print(i)
```

```python
for i in range(-3, 5):
    print(i)
```

```python
# third argument in range function
for i in range(2, 15, 2): # the range start with 2, add up 2 in every loops and stop before 15
    print(i)
```

```python
for i in range(1, 20, 3): # the range start with 1, add up 3 in every loops and stop before 20
    print(i)
```

```python
# range function only accept integers arguments
for i in range(3.14): # will return error
    print(i)
```

```python
# use for _ in range to execute the code specify number of times.
# example: print the message Python is great 7 times
for _ in range(7): # use _ to ignore the value produced by the range function in the for loops 
    print("Python is great!")
```
