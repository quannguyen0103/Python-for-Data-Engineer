```python
#set up different data types
# Global variables
int_var = 5
float_var = 5.3
string_var = "Hi Python"
bool_var = True
```

```python
#define new function: reassign all the arguments input in function
def reassign_arguments(int_var, float_var, string_var, bool_var):
    int_var = 25 #reassign values
    float_var = 34.6 #the varibale in here is local variables so it will not affect the original variables
    string_var = "Not Hello"
    bool_var = False
    # we can use different name, this example use these names to make us understand global vs local variable
    print("inside this function:", int_var, float_var, string_var, bool_var)
```

```python
#print out function result
reassign_arguments(int_var, float_var, string_var, bool_var)
```

```python
#print out the variables at beginning
print(int_var, float_var, string_var, bool_var) #the variable hasn't been changed
```
