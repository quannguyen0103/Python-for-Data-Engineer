```python
#Data wrangling: process of transform data from raw to another format, make it more valuable
# Concatenating datasets
# Your data can be broken chunks from multiple sources, so you need to concatenating (linking) them together
```

```python
#import pandas
import pandas as pd
# creating a dummy dataframe

df1_dummy = {
    "serial_id" : ["1", "2", "3", "4", "5"],
    "sale_month" : ["Jan", "Feb", "Mar", "Apr", "May"],
    "sales" : ["12300", "25100", "17800", "20100", "21000"]
}
#create dateframe from dictionary df1_dummy
df1 = pd.DataFrame(df1_dummy, columns = ["serial_id", "sale_month", "sales"]) #you can change column names

df1
```

```python
# use iloc to select rows in dataframe
df1.iloc[[0]]
```

```python
# use [] to select column from dataframe
df1["sale_month"].iloc[1]
```

```python
#create 2nd dummy table

df2_dummy = {
    "serial_id" : ["6", "7", "8", "9", "10"],
    "sale_month" : ["Jun", "Jul", "Aug", "Sep", "Oct"],
    "sales" : ["25000", "23700", "24600", "24000", "23950"]
}

df2 = pd.DataFrame(df2_dummy, columns = ["serial_id", "sale_month", "sales"])
```

```python
df2
```

```python
#create 3rd dummy table

df3_dummy = {
    "sales_threshold" : ["No", "Yes", "No", "Yes", "Yes", "Yes", "Yes", "Yes", "Yes", "Yes"],
    "bonus_threshold" : ["No", "Yes", "No", "No", "No", "Yes", "No", "Yes", "Yes", "No"]
}

df3 = pd.DataFrame(df3_dummy) #column names automatic assgin for dataframe
```

```python
df3
```

```python
# concat the first and second dataframe together row_wise
df_row = pd.concat([df1, df2])
# You can only concate columns row-wise with same names
```

```python
df_row #now 2 dataframe has been appended
```

```python
# concat df_row and 3rd dataframe column-wise
df_full = pd.concat([df_row, df3], axis = 1) #default is axis = 0, which means concat row-wise
df_full
# because the index in df_row is not unique, it can't concat with df3
```

```python
# to solve this, add an argument to ignore index
df_row = pd.concat([df1, df2], ignore_index = True)
```

```python
df_row
```

```python
# Now the index is unique, you can concat with 3rd dataframe
df_full = pd.concat([df_row, df3], axis = 1)
```
