```python
import pandas as pd
```

```python
data = pd.read_csv("artwork_data.csv", low_memory = False)
```

```python
data.head()
```

```python
#filter to get values with white space at the end of string
data.loc[data.title.str.contains("\s$", regex = True)]
```

```python
#use strip method to strip all white space at the end of string
data.title.str.strip()
```

```python
#string method don't take inplace arguments
data.loc[data.title.str.contains("\s$", regex = True)] #white space still hasn't remove
```

```python
# permanently remove white space
data.title = data.title.str.strip()
```

```python
data.loc[data.title.str.contains("\s$", regex = True)] #now all value with white space are gone
```

```python
# we can transform columns
data.title = data.title.transform(lambda x: x.strip())
```
