```python
if 30 < 56: # because the condition 30 < 56 is True, the statement within if block will be executed
    print("30 is less than 56")
```

```python
if 30 > 45: # because the condition 30 > 45 is False, the statement within if block will not be executed
    print("30 is bigger than 45")
```

```python
if 35 == 23: # compare
    print("well")
```

```python
if not 35 == 23: # not == equal to !=
    print("35 is not equal to 23")
```

```python
if not 35 = 23: # the = is to assign value, not to compare
    print("35 is not equal to 23")
```

```python
num = 23

if num > 0:
    print("the number is bigger than zero:", num) #Python will print 2 messages with space between
```

```python
if num: #automatic check if num is > 0
    print("the number is bigger than zero:", num)
```
