```python
popular_books = ["dac nhan tam",
                "tony buoi toi",
                "yeu than ki",
                "homo sapiens",
                 "crazy chick"]
```

```python
def modify_list(books):
    books.append("lean analytics")
```

```python
# now the function actually change the original variable, because append function permanently modify list
modify_list(popular_books) # the list has been update, it will append more if you keep run this code
print(popular_books)
```

```python
#more example
# create a function that will modify list
def modify_list(books):
    books[-3] = "hinh hoc thong minh" 
    books.pop()
    books.insert(2, "one punch man")
```

```python
# invoke the function
modify_list(popular_books)
print(popular_books)
```
