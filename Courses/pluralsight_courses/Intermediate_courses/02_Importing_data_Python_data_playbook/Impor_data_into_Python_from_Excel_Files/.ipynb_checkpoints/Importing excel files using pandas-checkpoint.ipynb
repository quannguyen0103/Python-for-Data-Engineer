```python
#first import pandas
import pandas as pd
```

```python
#load data from excel file into Python
excel_file = pd.ExcelFile("stackoverflow.xlsx")
```

```python
#check the how many sheets there are in excel files
excel_file.sheet_names
```

```python
#turn excel_file into data frame using parse
excel_df = excel_file.parse()
```

```python
excel_df.head()
```

```python
# another way to change exceL_file into dataframe, but with parameters
header_field = header_fields = ["New_Id", "New_PostTypeId", "New_CreationDate", "New_Score", "New_ViewCount",
                 "New_LastActivityDate", "New_Title", "New_Tags", "New_AnswerCount",
                "New_CommentCount", "New_FavoriteCount", "New_CloseDate"]
#use names argument instead of header argument to assign header, header argument only take integer
excel_header = pd.read_excel("stackoverflow.xlsx", names = header_field)
```

```python
excel_header
```

```python
# Load every sheets into Python and change they to dictionary key-value pairs
excel_sheets = pd.read_excel("stackoverflow.xlsx", sheet_name = None)
```

```python
# now the every sheet is in dictionary key-value pairs
# extract sheet like extract value from dictionary using key - the name of the sheet
excel_sheets["Posts"].head()
```

```python
# Import not entire file but specific sheet
excel_new_sheet = pd.read_excel("stackoverflow.xlsx", sheet_name = "Users", usecols = "A:E", header = 1).head()
```

```python
excel_new_sheet
```

```python
#another way to select columns and new way to eliminate empty column
excel_users_sheet = pd.read_excel("stackoverflow.xlsx", sheet_name = "Users", usecols = range(1,9), header = 1).head()
```

```python
excel_users_sheet
```

```python
#import the excel sheet by using index, not name
excel_users_sheet = pd.read_excel("stackoverflow.xlsx",
                                  sheet_name = 2,
                                  header = 1,
                                  usecols = range(1,9)).head()
```

```python
excel_users_sheet
```

```python
#skip rows: skip the number of rows, not the index of rows
excel_users_sheet = pd.read_excel("stackoverflow.xlsx",
                                  sheet_name = 2,
                                  header = 1,
                                  usecols = range(1,9),
                                 skiprows = 2).head()
```

```python
excel_users_sheet
```

```python
# nrows: specify how many rows to get
excel_worksheet = pd.read_excel("stackoverflow.xlsx",
                                  sheet_name = 2,
                                  header = 1,
                                  usecols = range(1,9),
                               nrows = 2)
```

```python
excel_worksheet
```

```python
#change the type of columns using dtype argument
pd.read_excel("stackoverflow.xlsx",
              sheet_name = 0,
              header = None,
              usecols = range(1,9),
              nrows = 10,
              dtype = {"Location" : str, "Reputation" : int},
             parse_dates = True)
#parse_dates = True: automatic turn every datetime column into datetime type
#parse_dates = ["column_name"] : turn a specific columns into datetime
```

```python
#converters: apply function for particular columns
pd.read_excel("stackoverflow.xlsx",
              sheet_name = 2,
              header = 1,
              #use a anonymous function, lambda to add 100 for every Id
             converters = {"Id" : lambda x: x + 100})
```

```python
#replace NaN with empty values with keep_default_na argument
pd.read_excel("stackoverflow.xlsx",
              sheet_name = 0,
              names = header_field,
              usecols = range(10),
              keep_default_na = False).head(5)
```
