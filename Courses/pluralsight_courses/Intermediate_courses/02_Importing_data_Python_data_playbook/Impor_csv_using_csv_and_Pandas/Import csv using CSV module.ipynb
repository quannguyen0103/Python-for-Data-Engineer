```python
import csv
```

```python
#open the csv file
with open("users-simple-five.csv", "r") as csv_file:
    #reader function: return a reader object that can used to iterate over the lines
    csv_reader = csv.reader(csv_file, delimiter =",", quotechar = '"')
    for row in csv_reader: #print all row in csv_reader file
        #each row will be list
        print(row)
```

```python
#DictReader function: return a reader object, but in dictionary format
#Specify field names
field_names = ["Id", "Reputation", "Location", "DisplayName"]
with open("users-simple-five.csv", "r") as csv_file:
    csv_dict_reader = csv.DictReader(csv_file, fieldnames = field_names)
    #fieldnames = field_names: specify field_name for dict reader object
    for row in csv_dict_reader:
        # each row is a list so we can subset values in a row
        print(row["DisplayName"] + " has a reputation of " + row["Reputation"])
```

```python
#Load unquote value as numberic value
with open("users-simple-five.csv", "r") as csv_file:
    csv_reader = csv.reader(csv_file, quoting = csv.QUOTE_NONNUMERIC)
    #csv.QUOTE_NONNUMERIC: turn unquote value to numberic data type
    for row in csv_reader:
        #each row will be a list
        print(row)
```
