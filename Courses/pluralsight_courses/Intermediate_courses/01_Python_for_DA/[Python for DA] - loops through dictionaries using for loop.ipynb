```python
names_and_ages = {"Quan" : 25,
                "Quyen" : 33,
                "Quy" : 29,
                "Duong" : 33}
```

```python
#iterating over dictionary keys
for key in names_and_ages:
    print(key)
```

```python
#iterating over dictionary keys and values
for key in names_and_ages:
    print("Name:", key, "\nAge:", names_and_ages[key], "\n")
```

```python
#iterating over dictinary keys
for key in names_and_ages.keys():
    print(key)
    
print("\n")

#iterating over dictinary values
for value in set(names_and_ages.values()): # use set() to return only unique values
    print(value)
```

```python
#return every keys-values pairs using items() function
for key, value in names_and_ages.items():
    print(key, value)
    
print("\n")
#return every keys-values pairs as tuple form using items() function
for item in names_and_ages.items():
    print(item)
```
