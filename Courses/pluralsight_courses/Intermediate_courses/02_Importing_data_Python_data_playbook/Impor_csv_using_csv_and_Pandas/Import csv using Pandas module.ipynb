```python
# Option 2: Provide column names
#1. Create a header field list
header_fields = ["New_Id", "New_PostTypeId", "New_CreationDate", "New_Score", "New_ViewCount",
                 "New_LastActivityDate", "New_Title", "New_Tags", "New_AnswerCount",
                "New_CommentCount", "New_FavoriteCount", "New_CloseDate"]
```

```python
# assign header for dataframe using names argument
file_header = pd.read_csv(remote_file, names = header_fields)
```

```python
file_header.head() #now the dataframe has headers
```

```python
#dtypes functions: check the data type of columns
pd.read_csv("posts-100-header.csv", usecols = [0, 5, 8, 9]).dtypes
```

```python
# We can use dtypes to change the data type in columns
pd.read_csv("posts-100-header.csv", usecols = [0, 5, 8, 9], dtype = {"Id": float}).dtypes
```

```python
# Apply function to specific columns
post_tags = pd.read_csv("posts-100-header.csv", usecols = [0, 5, 8, 9])
post_tags.iloc[1] #iloc[]: get the row by index
```

```python
import re #the module that contain functions using for apply function to columns
```

```python
# Turn a string to a list of string
post_tags = pd.read_csv("posts-100-header.csv", usecols = [0, 1, 2, 7],
                       converters = {"Tags": lambda x: re.findall("<[A-Za-z0-9_-]*>", x)})
#This is very powerful
```

```python
post_tags.head() #now the tags contain list of strings 
```

```python
# date and time in Pandas using parse_dates
post_tags = pd.read_csv("posts-100-header.csv", usecols = [0, 1, 2, 7],
                        parse_dates = ["CreationDate"])
```

```python
post_tags.dtypes #now the creationdate is datetime type
```

```python
#Missing value
pd.read_csv("posts-100-header.csv", usecols = [0, 3, 4, 8, 9, 10]).head()
```

```python
# NaN repesent missing values
# Work with NaN
pd.read_csv("posts-100-header.csv", usecols = [0, 3, 4, 8, 9, 10], na_filter = False).head()
#Now all NaN will be replace with empty value
```
