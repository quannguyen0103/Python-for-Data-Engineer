```python
#open a file like this and you won't need to close file afterward
car_specs = []

#with block below is ensure the file is close when you exist the directory
with open ("./files/car_hp.txt", "r") as file_again:
    
    for car in file_again: #loop and append each value from the file to car_specs list
        car_specs.append(car.rstrip("\n")) #strip the car item, add lines
```

```python
car_specs #now the list contain all content from the file
```

```python
file_again.read() #the file_again has been close with the with block above so we can't read 
```

```python
# to append a value to a file, first open file in append mode
with open ("./files/car_hp.txt", "a") as f: # "a": append mode
    more_cars = f.write("\nBuick")
    
    print("Number of character:", more_cars)
```

```python
!cat ./files/car_hp.txt #take a look at new file
```

```python
#write mode
sport_cars = ["Lambo", "Chiron", "Regera", "Ferrari"]

with open ("./files/car_hp.txt", "w") as f: #"w": write mode
    for car in sport_cars:
        #write mode will destroy all the existing content and replace with new content
        f.write(car)
```

```python
!cat ./files/car_hp.txt #now all the cars in sport_cars has replace existing cars
```

```python
with open ("./files/car_hp.txt", "w") as f:
    for car in sport_cars:
        f.write(car + "\n") #add empty line between each value
```
