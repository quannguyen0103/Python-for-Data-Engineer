```python
def calculate_sum(a, b, c): #define a function with input arguments
    #1
    result = a + b + c #calculate the result
    #2
    print("The sum of a, b, c is", result) #print result
    #3
    if result < 10: # check if the result is < 10, if True the if statement will be executed
        print("Result is smaller than 10!")
    else:
        print("Result is bigger than 10!")
```

```python
#the calculate_sum(a, b, c) expect 3 arguments
calculate_sum(2) #this will not work
```

```python
#this will work
calculate_sum(2, 5, 7)
```

```python
# this will also work
x = 25
y = 13
z = 26
calculate_sum(x, y, z)
```
