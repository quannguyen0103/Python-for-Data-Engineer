```python
#break: use to control while loops and for loops execution
classic_cars = ["mustang", "grey", "suck", "gaygale", "dogsweet"]
for car in classic_cars:
    print(car)
    break # when the first loop executed, it will break so only first value is printed
```

```python
for car in classic_cars:
    print(car.title())
    #if block is still inside for loops, so it will iterate every car value along side print statement above
    if car == "gaygale":
        print( car.title(), "is super expensive!")
        break # when car == "gaygale", the break will stop the loop. Nothing will be printed after gaygale
    # when if statement is True, every code after break will not be printed
    # (only in the loop where if condition is True)
    print("still within the loop:", car)
```

```python
num_list = [100, 23, 456, 234, 45, 34, 67, 53, 22]
for num in num_list: #the loops will execute code in order
    #1
    print("check....", num)
    #2
    if num % 3 == 0:
        print(num, "can multiple by 3!")
        break # when a number in num_list % 3 == 0 appears, it will stop the loops
    #3
    print(num, "can't multiple by 3 :(") #when if condition still False, it will still executed this code
```
