```python
#normalize data: transform number data into range 0-1 to apply machine learning on it
# There are 3 way to normalize data
```

```python
import pandas as pd
```

```python
#read dataset
wine_data = pd.read_csv("winequality-red.csv", sep = ";") # data separate in csv file by the ';'
```

```python
wine_data.head()
```

```python
#1. Simple featuring scaling
#formular: X_new = X / Xmax
#apply on specific column
wine_data["fixed acidity"] = wine_data["fixed acidity"] / wine_data["fixed acidity"].max()
wine_data
```

```python
#apply on all dataset
wine_data = wine_data/wine_data.max()
wine_data
```

```python
#2. Min- max scaling
#formular: Xnew = (X - Xmin)/(Xmax - Xmin)
# we can use the eaiser method: using scikit-learn - all important machine learning library
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
#select all columns in dataframe and transform data
wine_data[wine_data.columns] = scaler.fit_transform(wine_data[wine_data.columns])
```

```python
wine_data
```

```python
#3. Z-score normalization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

wine_data[wine_data.columns] = scaler.fit_transform(wine_data[wine_data.columns])
```
