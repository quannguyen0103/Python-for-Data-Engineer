```python
#importing necessary library
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
#read csv file
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
#check how many unique values in a column
titanic["Pclass"].nunique()
```

```python
#get the number of people (Survived = 1 is survived person)
titanic["Survived"].sum()
```

```python
# use groupby function to group data into group base on Pclass
titanic_class = titanic.groupby("Pclass")
```

```python
# only get the group of Pclass = 1
titanic_class.get_group(2) # 2 here is value in titanic Pclass, not index
```

```python
#Aggregate
```

```python
# how many people survive per class
titanic_class["Survived"].sum()
```

```python
# average age of people from different class
titanic_class["Age"].mean()
```

```python
# max ticket fare paid
titanic_class.max()
```

```python
# multiple aggregate at once
titanic_class.agg({"Fare": ["sum", "count", "max"]})
```

```python
#Filter data with group by: take only parts of the data base on condition
#Transform data with group by: transfrom full set of data into another form
```

```python
#1. Filter
```

```python
# What is the average fare when a group has its average age greater than 38?
titanic_class.filter(lambda x: x["Age"].mean() > 38.0)
```

```python
titanic_class.filter(lambda x: x["Age"].mean() < 38.0)
```

```python
# Get the average fare for group with age > 38 anf group with age < 38
print(titanic_class.filter(lambda x: x["Age"].mean() > 38.0)["Fare"].mean())
print(titanic_class.filter(lambda x: x["Age"].mean() < 38.0)["Fare"].mean())
```

```python
# Transfom data
```

```python
# Which of the passengers paid a fare that is higher than the average for their ticket class?
# create a new column name average_fare_class to set the avarage fare of each class for every guests
titanic["average_fare_class"] = titanic.groupby("Pclass")["Fare"].transform(lambda x: x.mean())
```

```python
titanic
```

```python
#create a new column, assign True if fare above average or False if vice versa
titanic["fare_above_avg"] = titanic["Fare"] > titanic["average_fare_class"]
```

```python
#count how many True and False in column fare_above_avg
titanic_fare = titanic.groupby("fare_above_avg").count()
```

```python
titanic_fare
```
