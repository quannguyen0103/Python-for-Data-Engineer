```python
# gender of a typical Nobel Prize winner?
#1. Import dataset
#2. Clean and transform data
#3. Plot the data
```

```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#1. Import dataset
data = pd.read_csv("dataset.csv")
data.head()
```

```python
#2. Clean and transform data
#get the number of each gender
gender = data.groupby("sex").size().reset_index(name = "number") #count values in each gender

# Calculate the sum of total winners
total_gender = data["sex"].count()

#Calculate the proportion of each gender
gender_prop = gender["number"]/total_gender
gender_prop
```

```python
#3. Plot the data

#create labels
labels = ["Female", "Male"]
colors = ["lightgrey", "darkorange"]

# create a pie chart
plt.pie(gender_prop,
       labels = labels, #name of each part
       autopct='%1.1f%%', #add percentage
       colors = colors, #change color
       startangle = 90) #rotate the pie chart

#change plot tile
plt.title("Nobel-Prize-Winner gender proportion")
```
