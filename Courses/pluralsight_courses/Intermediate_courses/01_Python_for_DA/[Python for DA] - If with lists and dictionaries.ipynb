```python
#set up variables
classic_cars = ["mustang", "grey", "suck"]
sport_cars = ["lambo", "i8", "ferrari"]
plane = ["a370", "747", "a380"]
```

```python
if len(classic_cars) < 5:
    print("classic cars is not a lot")
```

```python
# check if a value is in a list
transportation = "lambo"

if transportation in classic_cars:
    print(transportation.capitalize(), "is a classic car")

if transportation in sport_cars:
    print(transportation.capitalize(), "is a sport car")

if transportation in plane:
    print(transportation.capitalize(), "is a plane")
```

```python
# check if a key is in a dictionary
cars_and_prices = {"lambo" : 300000,
                  "farrari": 200000,
                  "hondai": 10000}

if "farrari" in cars_and_prices:
    print("the price of", "farrari".capitalize(), "is", cars_and_prices["farrari"], "dollars")
```

```python
# check if a key is not in a dictionary
cars_and_prices = {"lambo" : 300000,
                  "farrari": 200000,
                  "hondai": 10000}

if "maybach" not in cars_and_prices:
    print("maybach".title(), "is not in the dictionary")
```
