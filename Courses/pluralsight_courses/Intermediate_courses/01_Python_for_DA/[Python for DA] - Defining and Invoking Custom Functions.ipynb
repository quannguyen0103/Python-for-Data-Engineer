```python
# another example of reuse function using variables
# set up value a, b, c
a = 4
b = 2
c = 3

def calculate_sum(): #define a function, the order of code being executed is below
    #1
    result = a + b + c #calculate the result
    #2
    print("The sum of a, b, c is", result) #print result
    #3
    if result < 10: # check if the result is < 10, if True the if statement will be executed
        print("Result is smaller than 10!")
```

```python
calculate_sum()
```

```python
#update value of a and b
a = 1
b = 7
calculate_sum() #this function will be updated too
```
