```python
#define a function
def power_number(a, b):
    power = a ** b
    print(power)
```

```python
#invoke power number
power_number(3, 5) # = 3 **5
```

```python
#you can store function result like this
result = power_number(3, 5)
```

```python
print(result) #result is none
```

```python
# how to store function result
def power_number(a, b):
    power = a ** b
    return power #the return statement will return the result, not printed
```

```python
#now you can store function result
result = power_number(2, 4)
print(result)
```

```python
# a function can take in and return any type of data
def square_and_cube_and_divide(num):
    square = num ** 2
    cube = num ** 3
    divide = num / 2
    return square, cube, divide #return a tuple of results
```

```python
# invoke a function and store result
output = square_and_cube_and_divide(6)
print(output) 
```

```python
# store result in different variables
square, cube, divide = square_and_cube_and_divide(12) #each value of tuple will assign 
```

```python
print(square)
```

```python
print(cube)
```

```python
print(divide)
```

```python
# make sure when create a function, every outcome will return
def divisible_check(a, b):
    
    if a % b == 0:
        return str(a) + " is divisible by " + str(b)
    elif a % b != 0:
        return str(a) + " is not divisible by " + str(b)
```

```python
print(divisible_check(9, 3))
```

```python
print(divisible_check(9, 2))
```
