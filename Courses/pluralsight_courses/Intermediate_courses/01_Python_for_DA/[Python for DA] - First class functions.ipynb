```python
# function can take in arguments from other functions, and can also return values from other functions
# create a function to calculate GST (a type of tax)
def cal_GST(price, type_of_item):
    
    if type_of_item == "product":
        return price * 1.05
    
    elif type_of_item == "service":
        return price * 1.1
    
    else:
        return price * 1.3
```

```python
# create a function calculate VAT using cal_price arguments
def cal_VAT(price, type_of_item): # VAT is a type of tax
    return price * 1.2 #VAT always 20%, no matter what type of item
```

```python
type_of_tax = cal_VAT
type_of_tax(1000, "food")
```

```python
type_of_tax = cal_GST
type_of_tax(1000, "food")
```

```python
# create a function that return another function
def find_tax_calculator(type_of_tax):
    
    if type_of_tax == "GST":
        return cal_GST #cal_GST is a function
    
    elif type_of_tax == "VAT":
        return cal_VAT#cal_VAT is a function
```

```python
#assign new variable for function
GST_tax = find_tax_calculator("GST")
# now we can use GST_tax like cal_GST
GST_tax(1200, "product")
```

```python
# we can use function as argument for another function
def get_price(price, type_of_product, tax = GST_tax):
    return tax(price, type_of_product)
```

```python
#invoke get price function, tax argurment has been define as default (GST_tax) so we don't need to input
get_price(1500, "service")
```
