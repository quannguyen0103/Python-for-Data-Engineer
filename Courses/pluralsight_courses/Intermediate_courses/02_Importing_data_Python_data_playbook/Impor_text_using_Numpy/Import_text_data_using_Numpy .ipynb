```python
#numeric txt file with comma
badges_comma = np.loadtxt("badges-five.txt")
#because data seperate by the comma look like float, python can not specify it
```

```python
#You have to specify that the separate is the comma
badges_comma = np.loadtxt("badges-five.txt", delimiter = ",")
```

```python
badges_comma #now it works
```

```python
#Load a subset of columns
badges_comma = np.loadtxt("badges-five.txt", delimiter = ",", usecols = (0,2))
#use only the first and third columns
```

```python
badges_comma
```

```python
#If the numeric five has header, we can use skiprow to skip the first row
badges_header = np.loadtxt("badges-five-header.txt", delimiter =",", skiprows = 1)
```

```python
badges_header
```

```python
#dtype: check the type of data in files
badges_header.dtype
```

```python
#add a particular value to columns
def increase(the_id):
    return int(the_id) + 1000 #add index start with 1000 for every columns
badges_id = np.loadtxt("badges-five.txt", delimiter = ",", dtype = int, converters = {0: increase}) 
```

```python
badges_id
```

```python
#genfromtxt() function: use to deal with missing values
badges_missing_values = np.genfromtxt("badges-five-missing-value.txt", delimiter = ",")
```

```python
badges_missing_values #missing value or not numeric data value now become nan 
```

```python
badges_missing_values = np.genfromtxt("badges-five-missing-value.txt", delimiter = ",", skip_header = 1)
```
