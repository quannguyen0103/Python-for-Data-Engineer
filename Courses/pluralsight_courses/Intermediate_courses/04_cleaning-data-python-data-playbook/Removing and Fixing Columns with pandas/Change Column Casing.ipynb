```python
import pandas as pd
```

```python
data = pd.read_csv("artwork_sample.csv")
```

```python
data.columns #get all the column names
```

```python
#change all column name to lower case
data.columns.str.lower()
```

```python
#you can lower loop through the list of column names
[x.lower() for x in data.columns] #data.columns is a list so we can loop through it
```

```python
#note: the original column names still unaffect
data.columns
```

```python
#if you want to change it permantly, assign the columns to the new array
data.columns = [x.lower() for x in data.columns]
```

```python
data.columns #now it is all lower case
```

```python
#if you want to change it back, you can import the csv file again
data = pd.read_csv("artwork_sample.csv")
```

```python
data.columns
```

```python
#we can also use map instead of the for loops
data.columns = map(lambda x: x.lower(), data.columns) #map the function to every column names
```

```python
data.columns
```

```python
import re
```

```python
#lower every columns and add _ for every name with 2 characters
data = pd.read_csv("artwork_sample.csv")
data.columns = [re.sub(r"([A-Z])", r"_\1", x).lower() for x in data.columns]
#we add _ after the uppcase letter, and then lower every names
```

```python
data.columns #now it's in the same format
```
