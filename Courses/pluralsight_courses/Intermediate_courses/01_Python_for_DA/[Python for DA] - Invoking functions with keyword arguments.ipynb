```python
def car_parts(name, brand, hp, engine, zero_to_sixty, top_speed, price):
    
    other_details = hp, engine,zero_to_sixty, top_speed
    
    print("Name of the car:", name.title())
    print("Brand of the car:", brand.title())
    print("Price of the car:", price)
    print("Other details (hp, engine, zero_to_sixty, top_speed):", other_details)
```

```python
#Invoke a function like above can be easily be mistype or miss arguments
car_parts("kia morning", "honda", 544, 7.0, 2.9, 220, 120000)
```

```python
# Better way to invoke long function is specify the input arguments, rather than use index
car_parts(brand = "honda",
          name = "kia morning",
          zero_to_sixty = 2.9,
          engine = 7.0,
          hp = 544,
          top_speed = 220,
          price =  120000)
# This code will print the same result as above, despite we change the position of arguments
# Write code like this is longer, but far more easier to read
```

```python
# assign default value for arguments

#assign default value for price
def car_parts(name, brand, hp, engine, zero_to_sixty, top_speed, price = 150000):
    
    other_details = hp, engine,zero_to_sixty, top_speed
    
    print("Name of the car:", name.title())
    print("Brand of the car:", brand.title())
    print("Price of the car:", price)
    print("Other details (hp, engine, zero_to_sixty, top_speed):", other_details)
```

```python
# now that we assign default value for price, we don't need to input price argument when invoke function
car_parts(brand = "honda",
          name = "kia morning",
          zero_to_sixty = 2.9,
          engine = 7.0,
          hp = 544,
          top_speed = 220)
```

```python
# we can overwrite the default price
car_parts(brand = "honda",
          name = "kia morning",
          zero_to_sixty = 2.9,
          engine = 7.0,
          hp = 544,
          top_speed = 220,
          price =  200000)
```

```python
# we can create a function with all default value
def car_parts(brand = "honda",
          name = "kia morning",
          zero_to_sixty = 2.9,
          engine = 7.0,
          hp = 544,
          top_speed = 220,
          price =  200000):
    
    other_details = hp, engine,zero_to_sixty, top_speed
    
    print("Name of the car:", name.title())
    print("Brand of the car:", brand.title())
    print("Price of the car:", price)
    print("Other details (hp, engine, zero_to_sixty, top_speed):", other_details)
```

```python
# now that we assign values for all argument, we don't need to input any arguments
car_parts()

print("\n") #empty line
# you can still overwrite the default value by continue using arguments
car_parts(name = "Toyota", price = "150000")
```
