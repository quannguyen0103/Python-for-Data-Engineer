```python
classic_cars =["mustang", "gaygale", "dogbird"]
```

```python
for car in classic_cars: # allow to execute the print(car) statement over and over again
    print(car)
```

```python
print(car) # the value car will be the last value in the list
```

```python
#loop through a string
for character in "Phantom":
    print(character)
```

```python
# number can't be iterated
for num in 2353436:
    print(num)
```

```python
#iterating through a tuple exactly like list
classic_cars =("mustang", "gaygale", "dogbird")

for car in classic_cars:
    print(car)
```

```python
# if else statements in for loop
num_list = [100, 200, 400, 235, 50, 600, 350]

for num in num_list:
    if num < 300:
        print("< 300:\n", num)
    else:
        print(">= 300:\n", num)
```
