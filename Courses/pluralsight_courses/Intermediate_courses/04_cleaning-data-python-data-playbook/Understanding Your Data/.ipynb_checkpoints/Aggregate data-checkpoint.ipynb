```python
import pandas as pd
```

```python
#import csv file and transform it into dataframe
data = pd.read_csv("artwork_sample.csv")
```

```python
#check data
data.head()
```

```python
#find the min of specific column
data.height.min()
```

```python
#find the max of specific column
data.height.max()
```

```python
#sum
data.height.sum()
```

```python
#average
data.height.mean()
```

```python
# median: number in the middle
data.height.median()
```

```python
#std
data.height.std()
```

```python
#agg function
data.agg("min") #return all the min value, even string column
```

```python
#the advantage of agg is that we can pass multiple aggregates to it
data.select_dtypes(include = "number").agg(["min", "max", "sum", "mean"]) #you need to put it on a list first
```

```python
#you can use the describe() function to get all the basic statistical values
data.describe()
```

```python
data.height.describe() #describe a column
```

```python
#turn a row into a column
data.select_dtypes(include = "number").agg("mean", axis = "columns")
#select_dtypes(include = "number"): select only column with numeric types data
```
