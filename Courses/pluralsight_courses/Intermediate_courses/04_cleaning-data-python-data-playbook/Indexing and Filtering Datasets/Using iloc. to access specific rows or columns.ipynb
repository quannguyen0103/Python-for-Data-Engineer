```python
import pandas as pd
```

```python
data = pd.read_csv("artwork_sample.csv")
```

```python
#lowercase all column names
data.columns = [x.lower() for x in data.columns]
```

```python
data.head()
```

```python
#iloc select rows base on their positions, not their index. loc is vice versa
data.iloc[:3, :] #not include row 3
```

```python
# the different between loc and iloc is that loc include last row, iloc is not
```

```python
#set the index to column id
data.set_index("id", inplace = True)
```

```python
#now index are id columns
data.head()
```

```python
#use iloc to select rows
data.iloc[:3, :]
```

```python
#use loc to select rows
data.loc[0:3, :]
#because loc look at the index, not the position of the rows so they need to access index 1035-1044
```

```python
# Use iloc is much more easier, and more Python logic (exclude the last row)
```

```python
# iloc also select columns base on position, not names
data.iloc[:3, :5] #select first 3 rows, first 5 columns
#because we set the index to id, the id columns is not a column now
```

```python
# we can reset index back
data.reset_index(inplace = True) #reset all index
```

```python
data.iloc[:3, :5] #now id back to a column
```

```python
#select specific rows and columns
data.iloc[[3,5], [2, 7]] #select row 3 and 5, column 2 and 7
```
