```python
# Introducing Pandas
import pandas as pd
```

```python
# DataFrame: an object that has rows and columns, use to hold data is table forms
# DataFrame is case sensitive
names_ages_genders = pd.DataFrame({
     "name": ["quan", "quy", "quyen", "duong"],
     "age": [25, 33, 35, 36],
     "gender": ["male", "female", "male", "female"]})
```

```python
names_ages_genders #output is tabular table form
```

```python
# describe function: give statistic overview of all the numeric columns in dataframe
names_ages_genders.describe()
```

```python
#subset columns in DataFrame
subset_table = names_ages_genders[["name", "age"]]
```

```python
subset_table
```

```python
# groupby function: group data to perform calculator
group_df = names_ages_genders.groupby(by = "gender").count()#count people in groups of gender
```
