```python
#set up variables
salary = 600
expenses = 300
```

```python
living_with_means = False

if salary - expenses > 0:
    living_with_means = True
```

```python
print("Am I living with means?", living_with_means)
```

```python
#complex if statement
if 100 > 50 > 10: # Python will first check if 100 > 50, then check if 50 > 10
    print("100 > 50 > 10")
    
if 100 < 50 < 10: # False
    print("100 < 50 < 10")
```

```python
# and operator
salary = 600
expenses = 300
living_with_means = salary - expenses > 0 #because living_with_means is > 0 so its value now is True

if living_with_means and expenses > 200: #living_with_means == True, expenses > 200 == True => overall True
    print("Can you lower your expenses?")
```

```python
# or operator
if 2 <= 5 or 2 > 3: # it will print the message if 1 of the statement are True
    print("this is correct")
```

```python
# You can have multiple lines of code in if block
salary = 600
expenses = 300

if salary > expenses:
    saving = salary - expenses #assign saving, you can't assign salary - expenses = saving
    print("This month i save", saving, "dollars!")
```

```python
# if else statement
salary = 600
expenses = 800
spend = salary - expenses

if salary > expenses: #
    print("This month i save", spend, "dollars!")
else:#because salary > expenses is False so this message will be print
    print("This month i overspend", abs(spend), "dollars!")
```

```python
# function in if else statement
num_list = [-2, 5, 6, 2, -10, -4]

if min(num_list) < 0: # check if smalles value in num_list is < 0
    # because the min(num_list) < 0 so the if statement will be executed
    print("smallest list value is smaller than 0")
else:
    print("smallest list value is bigger than 0")
```

```python
# more complex if else statements
salary = 500
expenses = 800

#set saving and overspend = 0 so when we print something outside of if block, it will print correct values
saving = 0
overspend = 0

if salary > expenses:
    saving = salary - expenses
    print("I save", saving, "dollars this month!")
else: # because salary < expenses => print in else statement will be executed
    overspend = abs(salary - expenses)
    print("I overspend", overspend, "dollars this month!")
    
print("saving:", saving, "overspend:", overspend) # outside of if blocks
```
