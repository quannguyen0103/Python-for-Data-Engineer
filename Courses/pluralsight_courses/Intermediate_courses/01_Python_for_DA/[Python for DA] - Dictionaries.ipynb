```python
#dictionary of names and ages
names_and_ages = {"Quan" : 25,
                 "Quy" : 29,
                 "Quyen": 32}
print(names_and_ages)
```

```python
#dictionary doesn't have index, but we can look up within Python dictionary using []
names_and_ages["Quan"]
```

```python
#keys() function: look up to all keys in dictionary
names_and_ages.keys()
```

```python
#values() function: access to all values in dictionary
names_and_ages.values()
```

```python
#items() function: return a list of tuples, where fields of tuple are key-value pairs
names_and_ages.items()
```

```python
#add new key-value pairs into a dictionary
names_and_ages["Duong"] = 33
print(names_and_ages)
```

```python
#update key-value pairs in a dictionary
names_and_ages["Quan"] = 13
print(names_and_ages)
```

```python
#pop() function: remove random key-value pairs, it will change the original dict permanently
names_and_ages = {"Quan" : 25, #the original dictionary has been change by pop, so recall it here
                 "Quy" : 29,
                 "Quyen": 32,
                 "Duong": 33}
names_and_ages.pop("Duong")
print(names_and_ages)
```

```python
#del function: permanently delete a key-value pairs
del names_and_ages["Quyen"]
names_and_ages
```

```python
#Dictionary can contain complex data types like list
occupations_and_names = {"DA": ["Quan", "Hoa"],
                        "DE": ["Huy", "Quan"]}
print(occupations_and_names)
```

```python
# Look up value in list in dictionary
occupations_and_names["DE"][0]
```
